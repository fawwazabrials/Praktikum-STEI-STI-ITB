```python
import pandas as pd
data = pd.read_csv('P05-03.csv')
```

```python
# 1. Histogram distribusi banyaknya review rumah yang disewakan di wilayah West Region dengan bins 20

data[data['neighbourhood_group'] == 'West Region']['number_of_reviews'].value_counts().plot(kind='hist', bins=20)
```

```python
# 2. Pie chart perbandingan banyaknya masing-masing neighbourhood dengan harga rumah per malam di atas 300

data[data['price'] > 300]['neighbourhood'].value_counts().plot(kind='pie')
```

```python
# 3. Scatter p;ot dengan long sebagai x dan lat sebagai y dari rumah yang ada di neigbourhood Kallang

data[data['neighbourhood'] == 'Kallang'].plot(kind='scatter', x='longitude', y='longitude')
```

```python
# 4. Bubble plot dengan long sebagai x, lat sebagai y, dan banyaknya review sebagai ukuran dari rumah dengan banyak hari ketersediaan rumah dalam satu tahun di atas 100

data[data['availability_365'] > 100].plot(kind='scatter', x='longitude', y='longitude', s='number_of_reviews')
```

```python
# 5. Stacked bar char dari rumah yang disewakan oleh Richards dengan neighbourhood sebagai x dan jenis penyediaan ruang sebagai stacked y

data[data['host_name'] == 'Richards'].groupby(['neighbourhood', 'room_type']).size().unstack().plot(kind='bar', stacked=True)
```
